<h2>Total Null Values in the orders table group by order status
    

approved       → 2 <br>
canceled       → 404 <br>
delivered      → 6 <br>
invoiced       → 266 <br>
processing     → 273 <br>
shipped        → 936 <br>
unavailable    → 2 <br>

```python
# Cancelled orders → drop the rows (they'll never be delivered)
# Shipped orders → fill with estimated delivery date
# Processing orders → fill with estimated delivery date
# approved orders → fill with estimated delivery date
# invoiced orders → fill with estimated delivery date 
# delivered → fill with estimated delivery date 
# unavailable → drop the rows

df3 = df3[~((df3['order_status'] == 'canceled') & (df3['order_delivered_timestamp'].isnull()))]

df3.loc[
    (df3['order_status'] == 'shipped') & (df3['order_delivered_timestamp'].isnull()),
    'order_delivered_timestamp'
] = df3['order_estimated_delivery_date']

df3.loc[
    (df3['order_status'] == 'processing') & (df3['order_delivered_timestamp'].isnull()),
    'order_delivered_timestamp'
] = df3['order_estimated_delivery_date']

fill_statuses = ['shipped', 'processing', 'approved', 'invoiced', 'delivered']

df3.loc[
    (df3['order_status'].isin(fill_statuses)) & (df3['order_delivered_timestamp'].isnull()),
    'order_delivered_timestamp'
] = df3['order_estimated_delivery_date']

drop_statuses = ['canceled', 'unavailable']

df3 = df3[~(
    (df3['order_status'].isin(drop_statuses)) & (df3['order_delivered_timestamp'].isnull())
)]
```
